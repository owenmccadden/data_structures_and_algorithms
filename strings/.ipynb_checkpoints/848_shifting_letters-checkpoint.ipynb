## My Solution
- works for smaller values of len(s) and len(shifts), but exceeds runtime limits for larger datasets
- O(n^2) + O(n)

```python
from collections import defaultdict

class Solution:
    def shiftingLetters(self, s, shifts):
        self.shift_counts = defaultdict(int)
        for i in range(len(shifts)):
            for j in range(len(s[:i+1])):
                self.shift_counts[j] += shifts[i]
        return self.shift(s, shifts)

    def shift(self, s, shifts):
        output = []
        for pos, shift in self.shift_counts.items():
            index = ord(s[pos]) + shift % 26
            if index > 122:
                index = 96 + (index % 122)
            output.append(chr(index))
        return ''.join(output)
```

```python
class Solution2:
    def shiftingLetters(self, s, shifts):
        alphabet = 'abcdefghijklmnopqrstuvwxyz'
        alphabet_dict = {'a':1,'b':2,'c':3,'d':4,'e':5,'f':6,'g':7,'h':8,'i':9,'j':10,'k':11,'l':12,'m':13,'n':14,'o':15,'p':16,'q':17,'r':18,'s':19,'t':20,'u':21,'v':22,'w':23,'x':24,'y':25,'z':26}
        starting_sum = sum(shifts)
        c = 0
        result = ''
        for i in range(len(shifts)):
            char_position = alphabet_dict[s[i]]
            char_position += (starting_sum - c)
            char_position %= 26
            result = result + alphabet[char_position-1]
            c += shifts[i]
        return result
```
